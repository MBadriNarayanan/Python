```python
def input_matrix(matrix,row,column) :
    
    print(" Creating The Matrix \n")
    
    i = 0
    
    while i < row :
        
        j = 0
        
        while j < column :
            
            matrix[i][j] = int(input(" Enter Value : "))
            
            j = j + 1
            
        i = i + 1
        
    print("\n")
```

```python
def display_matrix(matrix,row,column) :
    
    print(" Displaying The Matrix \n")
    
    i = 0
    
    while i < row :
        
        j = 0
        
        while j < column :
            
            print(" ",matrix[i][j],end = " ")
            
            j = j + 1
            
        print("\n")
            
        i = i + 1
        
    print("\n")
```

```python
def add_matrix(matrix1,matrix2,row1,column1,row2,column2) :
    
    matrix = [[0] * column1 for i in range(0,row1)]
    
    if row1 == row2 and column1 == column2 :
        
        print(" Matrix Addition Possible \n")
    
        i = 0

        while i < row1 :
            
            j = 0
            
            while j < column1 :
                
                matrix[i][j] = matrix1[i][j] + matrix2[i][j]
        
                j = j + 1
            
            i = i + 1
    
    else : 
        
        print(" Matrix Addition Not Possible \n")
        
    return matrix
```

```python
def subtract_matrix(matrix1,matrix2,row1,column1,row2,column2) :
    
    matrix = [[0] * column1 for i in range(0,row1)]
    
    if row1 == row2 and column1 == column2 :
        
        print(" Matrix Subtraction Possible \n")
    
        i = 0

        while i < row1 :
            
            j = 0
            
            while j < column1 :
                
                matrix[i][j] = matrix1[i][j] - matrix2[i][j]
        
                j = j + 1
            
            i = i + 1
    
    else : 
        
        print(" Matrix Subtraction Not Possible \n")
        
    return matrix
```

```python
def multiply_matrix(matrix1,matrix2,row1,column1,row2,column2) :
    
    matrix = [[0] * column2 for i in range(0,row1)]
    
    if column1 == row2  :
        
        print(" Matrix Multiplication Possible \n")
    
        i = 0

        while i < row1 :
            
            j = 0
            
            while j < column2 :
                
                k = 0
                
                while k < row2 :
                    
                    matrix[i][j] += matrix1[i][k] * matrix2[k][j]
                    
                    k = k + 1
        
                j = j + 1
            
            i = i + 1
    
    else : 
        
        print(" Matrix Multiplication Not Possible \n")
        
    return matrix
```

```python
def matrix_transpose(matrix,row,column) :
    
    matrixtranspose = [[0] * column for i in range(0,row)]
    
    i = 0
    
    while i < column :
        
        j = 0
        
        while j < row :
            
            matrixtranspose[i][j] = matrix[j][i]
            
            j = j + 1
            
        i = i + 1
        
    return matrixtranspose
```

```python
def matrix_create():
    
    row = int(input(" Enter Row : "))
    
    print("\n")
    
    column = int(input(" Enter Column : "))
    
    print("\n")
            
    matrix = [[0] * column for i in range(0,row)]
    
    return matrix,row,column
```

```python
def two_matrix_create() :
    
    row1 = int(input(" Enter Row 1: "))
    
    print("\n")
    
    column1 = int(input(" Enter Column 1: "))
    
    print("\n")
            
    matrix1 = [[0] * column1 for i in range(0,row1)]
    
    row2 = int(input(" Enter Row 2: "))
    
    print("\n")
    
    column2 = int(input(" Enter Column 2: "))
    
    print("\n")
            
    matrix2 = [[0] * column2 for i in range(0,row2)]
    
    return matrix1,row1,column1,matrix2,row2,column2
```

```python
def main() :
    
    print(" Matrix Operations \n\n")
    
    flag = 1
    
    while flag == 1 : 
        
        print(" 1. Input & Display \n")
        
        print(" 2. Addition \n")
        
        print(" 3. Subtraction \n")
        
        print(" 4. Multiplication \n")
        
        print(" 5. Transpose \n")
        
        print(" 6. Exit \n")
        
        option = int(input(" Enter Option : "))
        
        print("\n")
        
        if option == 1 :
            
            details = matrix_create()
            
            matrix = details[0]
            
            row = details[1]
            
            column = details[2]
            
            input_matrix(matrix,row,column)
            
            display(matrix,row,column)
            
        elif option == 2 :
            
            details = two_matrix_create()
            
            matrix1 = details[0]
            
            row1 = details[1]
            
            column1 = details[2]
            
            matrix2 = details[3]
            
            row2 =details[4]
            
            column2 = details[5]
            
            print(" Matrix 1 \n")
            
            input_matrix(matrix1,row1,column1)
            
            display_matrix(matrix1,row1,column1)
            
            print(" Matrix 2 \n")
            
            input_matrix(matrix2,row2,column2)
            
            display_matrix(matrix2,row2,column2)
            
            matrix3 = add_matrix(matrix1,matrix2,row1,column1,row2,column2)
            
            print(" Resultant Matrix \n")
            
            display_matrix(matrix3,row1,column1)
            
            
        elif option == 3 :
            
            details = two_matrix_create()
            
            matrix1 = details[0]
            
            row1 = details[1]
            
            column1 = details[2]
            
            matrix2 = details[3]
            
            row2 =details[4]
            
            column2 = details[5]
            
            print(" Matrix 1 \n")
            
            input_matrix(matrix1,row1,column1)
            
            display_matrix(matrix1,row1,column1)
            
            print(" Matrix 2 \n")
            
            input_matrix(matrix2,row2,column2)
            
            display_matrix(matrix2,row2,column2)
            
            matrix3 = subtract_matrix(matrix1,matrix2,row1,column1,row2,column2)
            
            print(" Resultant Matrix \n")
            
            display_matrix(matrix3,row1,column1)
            
            
        elif option == 4 :
            
            details = two_matrix_create()
            
            matrix1 = details[0]
            
            row1 = details[1]
            
            column1 = details[2]
            
            matrix2 = details[3]
            
            row2 =details[4]
            
            column2 = details[5]
            
            print(" Matrix 1 \n")
            
            input_matrix(matrix1,row1,column1)
            
            display_matrix(matrix1,row1,column1)
            
            print(" Matrix 2 \n")
            
            input_matrix(matrix2,row2,column2)
            
            display_matrix(matrix2,row2,column2)
            
            matrix3 = multiply_matrix(matrix1,matrix2,row1,column1,row2,column2)
            
            print(" Resultant Matrix \n")
            
            display_matrix(matrix3,row1,column2)
            
        elif option == 5 :
            
            details = matrix_create()
            
            matrix = details[0]
            
            row = details[1]
            
            column = details[2]
            
            input_matrix(matrix,row,column)
            
            display(matrix,row,column)
            
            matrix3 = matrix_transpose(matrix,row,column)
            
            display_matrix(matrix3,column,row)
            
        else :
            
            print(" Exit \n\n\n")
            
    
        flag = int(input(" Do You Want To Continue Type 0 Or 1 : "))
        
        print("\n")
         
    print("\n The End \n")
```
