```python
def write_file(filename) :
    
    f = open(filename,'r+')
    
    flag = 1
    
    while flag == 1 :
        
        word = input(" Enter A Word : ")
        
        f.write(word + " ")
        
        flag = int(input(" Do You Want To Continue Type 0 Or 1 : "))
        
        print("\n")
        
    f.close()
```

```python
def read_file(filename) :
    
    f = open(filename,'r+') 
    
    for word in f :
        
        print(word) 
        
    print("\n")
    
    f.close()
```

```python
def modify_file(filename) :
    
    s = ""
    
    f = open(filename,'r+')
    
    a = f.read()
    
    a = a.split()
    
    flag = 1
    
    while flag == 1 :
        
        givenword = input(" Enter Word To Be Modified : ")
    
        modifyword = input(" Enter The New Word : ")
        
        for i in range(len(a)) :
        
            if(a[i] == givenword) :
            
                a[i] = modifyword
            
        flag = int(input(" Do You Want To Continue Type 0 Or 1 : "))
        
        print('\n')
        
    for i in range(len(a)) :
        
        s = s + " " + a[i]
        
    f.close()
    
    f = open(filename,'w')
    
    f.write(s)
    
    print(s)
    
    print("\n")
```

```python
write_file(r'C:\Users\badri\Desktop\Python Lab\Demo.txt')

read_file(r'C:\Users\badri\Desktop\Python Lab\Demo.txt')

modify_file(r'C:\Users\badri\Desktop\Python Lab\Demo.txt')
```
